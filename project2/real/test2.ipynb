```python
# Author: <Friðrik Daði Smárason, Guðbjörn Ingi Torfason, Randver Thor Thorarinsson>
# Date: <07/03/2024>
# Project: <Project 2>
# Acknowledgements: <>

import statistics
# Festum heiti í listum
ALL_FILES = ['01uM_E6.txt','1uM_E9.txt','10uM_E10.txt','30uM_E4.txt','100uM_E7.txt','HOM_DMSO_D3.txt']
COLUMNS = ['AREA','AREACHANGE','ELONGATION','DISTANCE MOVED','VELOCITY']
# Skilgreinum fasta
SPACE = 15
PRINT_COLUMNS = len(ALL_FILES)+1 # +1 Til þess að reikna með bilinu sem hefur valkostinn og tölfræðiupplýsingar

# Fall til að sækja gögn um upplýsingar sem notandi vill
def get_data(file):
    data = open(file,'r')
    data_list = []
    for line in data:
        data_line = line.split(';') # Breyti setningunni í lista, skipt með ';'
        data_list.append(data_line) # Bæti við einni setningu í einu í listann.

    new_list = []
    for i in range(len(data_list)):
        # Athuga hvort fyrsta stak sé tala
        try:
            float(data_list[i][0])
            new_list.append(data_list[i])
        # Ef stak er ekki tala, hoppa yfir á næsta stak
        except:
            continue
    return new_list

# Fall til að breyta tölum í float og taka út gildi sem eru ekki mælingar, og reiknar út tölfræðiupplýsingar
def get_statistics(list,selection):
    wanted_data = []
    for i in range(len(list)):
        # Athuga hvort stakið sé tala
        try:
            a = float(list[i][selection])
            wanted_data.append(a)
        # Ef stak er ekki tala, hoppa yfir á næsta stak
        except:
            continue
    # Tölfræðireikningar
    MINIMUM = round(min(wanted_data),3)
    MAXIMUM = round(max(wanted_data),3)
    MEAN = round(statistics.mean(wanted_data),3)
    STDDEV = round(statistics.stdev(wanted_data),3)
    COUNT = len(wanted_data)
    # Stilli útkomunni út í vel skipulagðann lista
    DATA = [MINIMUM, MAXIMUM, MEAN, STDDEV, COUNT]
    return DATA

# Fall til að prenta eftir uppsetningu
def print_table(selection,in_list):
    print('-'*(SPACE*PRINT_COLUMNS))
    print(COLUMNS[selection-4].ljust(SPACE),ALL_FILES[0].replace('.txt','').rjust(SPACE),ALL_FILES[1].replace('.txt','').rjust(SPACE),ALL_FILES[2].replace('.txt','').rjust(SPACE),ALL_FILES[3].replace('.txt','').rjust(SPACE),ALL_FILES[4].replace('.txt','').rjust(SPACE),ALL_FILES[5].replace('.txt','').rjust(SPACE),sep='')
    print('-'*(SPACE*PRINT_COLUMNS))
    print('Minimum'.ljust(SPACE),str(in_list[0][0]).rjust(SPACE),str(in_list[1][0]).rjust(SPACE),str(in_list[2][0]).rjust(SPACE),str(in_list[3][0]).rjust(SPACE),str(in_list[4][0]).rjust(SPACE),str(in_list[5][0]).rjust(SPACE),sep='')
    print('Maximum'.ljust(SPACE),str(in_list[0][1]).rjust(SPACE),str(in_list[1][1]).rjust(SPACE),str(in_list[2][1]).rjust(SPACE),str(in_list[3][1]).rjust(SPACE),str(in_list[4][1]).rjust(SPACE),str(in_list[5][1]).rjust(SPACE),sep='')
    print('Mean'.ljust(SPACE),str(in_list[0][2]).rjust(SPACE),str(in_list[1][2]).rjust(SPACE),str(in_list[2][2]).rjust(SPACE),str(in_list[3][2]).rjust(SPACE),str(in_list[4][2]).rjust(SPACE),str(in_list[5][2]).rjust(SPACE),sep='')
    print('StdDev'.ljust(SPACE),str(in_list[0][3]).rjust(SPACE),str(in_list[1][3]).rjust(SPACE),str(in_list[2][3]).rjust(SPACE),str(in_list[3][3]).rjust(SPACE),str(in_list[4][3]).rjust(SPACE),str(in_list[5][3]).rjust(SPACE),sep='')
    print('Count'.ljust(SPACE),str(in_list[0][4]).rjust(SPACE),str(in_list[1][4]).rjust(SPACE),str(in_list[2][4]).rjust(SPACE),str(in_list[3][4]).rjust(SPACE),str(in_list[4][4]).rjust(SPACE),str(in_list[5][4]).rjust(SPACE),sep='')
    print('-'*(SPACE*PRINT_COLUMNS))

# Fall til að sýna notanda möguleika og biðja um val
def get_choice():
    print('Zebra fish analysis:')
    print('1: Area')
    print('2: Areachange')
    print('3: Elongation')
    print('4: Distance moved')
    print('5: Velocity')
    print('6: Quit')
    #selection = int(input('Selection: '))
    selection = 1
    return selection

# Í upphaf kóða, keyrt til að birta möguleika og biðja um val
selection = get_choice()

while selection != 6:
    selection += 3 # Af því að við viljum bara valkosti 4-8
    data_gathered = [] # Bý til tóman lista í byrjun til að fylla í
    for i in range(len(ALL_FILES)):
        data_cleaned = get_data(ALL_FILES[i]) # Sækir einungis þau gögn um tölfræði sem notandi vill
        data_gathered.append(get_statistics(data_cleaned,selection)) # Reiknar út tölfræði og tekur til
    print_table(selection,data_gathered) # Prentar útkomur
    #selection = get_choice() # Býður að velja aftur
    break
```
